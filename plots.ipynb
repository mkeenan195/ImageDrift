```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_month = ['March', 'April', 'August']

train_unstylized = [0.787, 0.533, 0.565]
train_low_style = [0.789, 0.541, 0.556]
train_high_style = [0.787, 0.527, 0.532]
```

```python
plt.figure(dpi=150)
plt.plot(test_month, train_unstylized, marker='d', linestyle='--')
plt.plot(test_month, train_low_style, marker='o', linestyle='--')
plt.plot(test_month, train_high_style, marker='*', linestyle='--')
plt.xlabel("Test Month")
plt.ylabel("mean Average Precision (mAP)")
plt.legend(['Trained on Unstylized (Original)', 'Trained on Low style', 'Trained on High style'])
plt.title('Variation of mAP with test months')
plt.show()
```

```python
test_month = ['March', 'April', 'August']

mar = [0.787, 0, 0]
apr = [0.412, 0.533,0]
aug = [0.427, 0.482, 0.565]

x = np.arange(len(test_month))  # the label locations
width = 0.3  # the width of the bars
labels = ['Jan to Feb', 'Jan to Mar', 'Jan to May']

fig, ax = plt.subplots(dpi=100)
rects1 = ax.bar(x, mar, width, label='March test images')
rects2 = ax.bar(x + width, apr, width, label='April test images')
rects3 = ax.bar(x + 2*width, aug, width, label='August test images')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('mean Average Precision (mAP)')
ax.set_xlabel('Training months')
ax.set_title('Plot showing the concept drift detection')
ax.set_xticks([.3, 1.45, 2.6], labels)
ax.set_ylim(0,1)
ax.legend()

fig.tight_layout()

plt.show()
```
